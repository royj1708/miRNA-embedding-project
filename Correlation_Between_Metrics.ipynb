```python
import pandas as pd


df = pd.read_csv('kidney_miRNA_pairs_pvalues_cossim_jaccard_correlation_emb.csv')
df['jaccard'] = pd.to_numeric(df['jaccard'], errors='coerce')
df['pvalue'] = pd.to_numeric(df['pvalue'], errors='coerce')
df['cos_sim'] = pd.to_numeric(df['cos_sim'], errors='coerce')
df['correlation'] = pd.to_numeric(df['correlation'], errors='coerce')
df['correlation_emb'] = pd.to_numeric(df['correlation_emb'], errors='coerce')

df = df[df['jaccard'].notna()]
df = df[df['pvalue'].notna()] 

print(df)
```

```python
from scipy.stats import pearsonr
corr1, _ = pearsonr(df["cos_sim"],df["pvalue"])
print('Pearsons correlation between cos_sim and pvalue: %.5f' % corr1)
corr2, _ = pearsonr(df["cos_sim"],df["jaccard"])
print('Pearsons correlation between cos_sim and jaccard: %.5f' % corr2)
corr3, _ = pearsonr(df["pvalue"],df["jaccard"])
print('Pearsons correlation between pvalue and jaccard: %.5f' % corr3)
```

```python
import matplotlib.pyplot as plt


plt.scatter(df["cos_sim"],df["pvalue"], color='k', s=25, marker="o")
plt.title('Correlation between cos_sim and pvalue', fontsize = 20)
plt.xlabel('cos_sim', fontsize = 20)
plt.ylabel('pvalue', fontsize = 20)
plt.legend()
plt.show()
```

```python
plt.scatter(df["cos_sim"],df["jaccard"], color='k', s=25, marker="o")
plt.title('Correlation between cos_sim and jaccard', fontsize = 20)
plt.xlabel('cos_sim', fontsize = 20)
plt.ylabel('jaccard', fontsize = 20)
plt.legend()
plt.show()
```

```python
plt.scatter(df["pvalue"],df["jaccard"], color='k', s=25, marker="o")
plt.title('Correlation between pvalue and jaccard', fontsize = 20)
plt.xlabel('pvalue', fontsize = 20)
plt.ylabel('jaccard', fontsize = 20)
plt.legend()
plt.show()
```
